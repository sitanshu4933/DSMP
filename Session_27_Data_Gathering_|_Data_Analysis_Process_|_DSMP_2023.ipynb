# Web Scrapping

```python
from bs4 import BeautifulSoup
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage=requests.get("https://www.ambitionbox.com/list-of-companies?campaign=homepage_companies_widget&page=2",headers=headers).text
```

```python
soup=BeautifulSoup(webpage,'lxml')
```

```python
company=soup.find_all('div',class_="company-content-wrapper")
```

```python
name=[]
rating=[]
review=[]
company_type=[]
headquarter=[]
old=[]
employees=[]
for i in company:
  name.append(i.find('h2').text.strip())
  rating.append(i.find('p',class_='rating').text.strip())
  review.append(i.find('a',class_='review-count').text.strip())
  company_type.append(i.find_all('p',class_='infoEntity')[0].text.strip())
  headquarter.append(i.find_all('p',class_='infoEntity')[1].text.strip())
  old.append(i.find_all('p',class_='infoEntity')[2].text.strip())
  employees.append(i.find_all('p',class_='infoEntity')[3].text.strip())
```

```python
d={'name':name,'rating':rating,'company_type':company_type,'headquarter':headquarter,'old':old,'employees':employees}
```

```python
df=pd.DataFrame(d)
```

```python
final_df=pd.DataFrame()
for j in (1,10):
  url="https://www.ambitionbox.com/list-of-companies?campaign=homepage_companies_widget&page={}".format(j)
  headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
  webpage=requests.get(url,headers=headers).text
  soup=BeautifulSoup(webpage,'lxml')
  company=soup.find_all('div',class_="company-content-wrapper")
  name=[]
  rating=[]
  review=[]
  company_type=[]
  headquarter=[]
  old=[]
  employees=[]
  for i in company:
    name.append(i.find('h2').text.strip())
    rating.append(i.find('p',class_='rating').text.strip())
    review.append(i.find('a',class_='review-count').text.strip())
    company_type.append(i.find_all('p',class_='infoEntity')[0].text.strip())
    headquarter.append(i.find_all('p',class_='infoEntity')[1].text.strip())
    old.append(i.find_all('p',class_='infoEntity')[2].text.strip())
    employees.append(i.find_all('p',class_='infoEntity')[3].text.strip())
  d={'name':name,'rating':rating,'company_type':company_type,'headquarter':headquarter,'old':old,'employees':employees}
  df=pd.DataFrame(d)
final_df=final_df.append(df)
final_df
```
